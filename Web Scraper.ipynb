```python
# Set up the environment

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import requests
from bs4 import BeautifulSoup
```

```python
#Get Chrome driver

driver = webdriver.Chrome()
driver.implicitly_wait(5)
```

```python
driver.get("https://www.expedia.com/things-to-do/search?%2Fthings-to-do%2Fsearch%3F&d1=2024-03-21&d2=2024-03-22&endDate=2024-03-22&filter.seeAll=true&location=Budapest%20%28and%20vicinity%29%2C%20Hungary&rid=179994&sort=RECOMMENDED&startDate=2024-03-21&swp=on")
time.sleep(5)

click_counter = 0

while click_counter < 6:
    try:
        
        # Find the "Show more" and click it
        show_more_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//*[@id="placeholder"]/div/button'))
        )
        show_more_button.click()
                
        # Sleep to reload page
        
        time.sleep(8)
        
        # Increase the counter
        click_counter += 1
        
    except Exception as e:
        print("An error occurred:", e)
        break
```

```python
# basic information 

# initialize dictionary with lists

df = {'Attraction_name':[], 'Rating':[], 'Ticket_price':[],'Free_cancellation':[],'Attraction_page':[]}

attractionname = driver.find_elements("xpath",'//*[@id="activity-list-4"]/div/section/div/a/span')
ratings = driver.find_elements("xpath",'//*[@id="activity-list-4"]/div/section/div/div[2]/section/div/div[2]/section/div[2]/div[1]')
tickets = driver.find_elements("xpath",'//*[@id="activity-list-4"]/div/section/div/div[2]/div/section/div[2]/div/section/span[2]')
freecancellations = driver.find_elements("xpath",'//*[@id="activity-list-4"]/div/section/div/div[2]/div/section/div[1]')
pages = driver.find_elements("xpath",'//*[@id="activity-list-4"]/div/section/div/a')

for attraction, rating, ticket, freecan, page in zip(attractionname, ratings, tickets, freecancellations, pages):
    df['Attraction_name'].append(attraction.get_attribute('textContent').strip())
    df['Rating'].append(rating.get_attribute('textContent').strip())
    df['Ticket_price'].append(ticket.get_attribute('textContent').strip())
    df['Free_cancellation'].append(freecan.get_attribute('textContent').strip())
    df['Attraction_page'].append(page.get_attribute('href'))
    
print(df)

#driver.close()
```

```python
# save the collected data in DataFrame

import pandas as pd

df_info = pd.DataFrame(df)
df_info
```

```python
# drop $ sign

df_info['Ticket_price'] = df_info['Ticket_price'].str.replace('[$,]', '', regex=True).astype(int)
df_info
```

```python
# replace values without free cancellation 

# Iterate over the Free_cancellation column and replace values not equal to "free cancellation available" with "No"
df_info['Free_cancellation'] = ['No' if value != 'Free cancellation available' else value for value in df_info['Free_cancellation']]
df_info
```

```python
# basic information

import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException

detailed = {"Attraction_name":[], "Travel_agency":[], "No.of_reviews":[], "Overview":[], "Attraction_page":[]}

for i in df_info["Attraction_page"][:10]:
    driver.get(i)
    
    try:
        attraction_name = driver.find_element("xpath", '//*[@id="overview"]/div/div/div/div[1]/h4').get_attribute('textContent').strip()
    except NoSuchElementException:
        attraction_name = '0'
        
    try:
        travel_agency = driver.find_element("xpath", '//*[@id="overview"]/div/div/div/div[1]/div[1]').get_attribute('textContent').strip()
    except NoSuchElementException:
        travel_agency = '0'
        
    try:
        no_of_reviews = driver.find_element("xpath", '//*[@id="overview"]/div/div/div/div[1]/div[2]/div/button').get_attribute('textContent').strip()
    except NoSuchElementException:
        no_of_reviews = '0'
        
    try:
        overview = driver.find_element("xpath", '//*[@id="overview"]/div/div/div/div[2]/div[2]/div').get_attribute('textContent').strip()
    except NoSuchElementException:
        overview = '0'

        
    detailed['Attraction_name'].append(attraction_name)
    detailed['Travel_agency'].append(travel_agency)
    detailed['No.of_reviews'].append(no_of_reviews)
    detailed['Overview'].append(overview)
    detailed['Attraction_page'].append(i)
    
    detailed_df = pd.DataFrame(detailed)
```

```python
#print results 

detailed_df
```

```python
# Remove "overview" from 'Overview' column

detailed_df['Overview'] = detailed_df['Overview'].str.replace('overview', '', case=False)
detailed_df['Travel_agency'] = detailed_df['Travel_agency'].str.replace('By', '', case=False)

detailed_df
```

```python
import re

detailed_df['No.of_reviews'] = detailed_df['No.of_reviews'].apply(lambda x: re.findall(r'\d+', x)[-1])
detailed_df
```
